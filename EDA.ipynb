**Hypothesis Test**
- H0: The average duration of rides from Loop to O'Hare is the same on rainy and non-rainy Saturdays.
- H1: The average duration of rides from Loop to O'Hare is different on rainy and non-rainy Saturdays.

Alpha set at 0.05 — appropriate for a non-critical domain. Using a two-sample t-test (Welch's) to compare means of two independent groups (rainy vs. non-rainy Saturdays), as it does not assume equal variances between groups.

```python
df_weather_conditions = pd.read_csv('data/weather_conditions.csv')
```

```python
df_weather_conditions.info()
```

```python
rainy = df_weather_conditions[df_weather_conditions['weather_conditions']
                              == 'Bad']['duration_seconds']
good = df_weather_conditions[df_weather_conditions['weather_conditions']
                             == 'Good']['duration_seconds']
```

```python
result = stats.ttest_ind(rainy, good)

print('p-value:', result.pvalue)
```

The p-value (6.52e-12) is significantly lower than the significance level of 0.05. Therefore, we reject the null hypothesis. There is statistically significant evidence that weather conditions affect the average duration of rides from Loop to O'Hare on Saturdays.
